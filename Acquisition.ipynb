```python
# create list of key terms to search
search_terms = ['litecoin']

# 2019-06-29
```

```python
# TWINT search
for i in search_terms:
    c = twint.Config()
    c.Search = i
    c.Min_likes = 0
    c.Since = '2019-01-01' #y-m-d
    c.Until = '2019-03-13'
    c.Store_csv = True
    c.Filter_retweets = True
    c.Output = f'{i}3.csv'
    c.Hide_output = True
    #c.Resume = 'Resume.txt'
    twint.run.Search(c)
```
